## Data Cleaning and Preprocessing

```python
## Import Pandas Library for Data Cleaning and Preprocessing
import pandas as pd
```

```python
## Load the data 
df = pd.read_csv("Medical_Appointment_No_Shows.csv")
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
## Drop the duplicate values according to dataset
df.drop_duplicates(["PatientId","No-show"]).head()
```

```python
## Check the statisics values of dataset
df.describe()
```

```python
## identify the age row which have minimum value is -1
mask=df.query("Age==""-1")
mask
```

```python
## drop that row in the dataset
df.drop(index=99832,inplace=True)
```

```python
## all the values are true for an appropriate dataset
df.describe()
```

```python
## Check the date and time format
print("ScheduledDay :", df['ScheduledDay'].head(1).values[0])
print("AppointmentDay sample:", df['AppointmentDay'].head(1).values[0])
```

```python
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'], format='%Y-%m-%dT%H:%M:%SZ')

df['ScheduledDay'].dt.strftime('%d-%m-%Y %H-%M')
```

```python
## print column name to check the cleaniness
print(df.columns)
```

```python
## Change the Hipertension : Hypertension and No-show : No_show
df.rename(columns={"Hipertension":"Hypertension"},inplace=True)
df.rename(columns={"No-show":"No_show"},inplace=True)
df.head()
```
